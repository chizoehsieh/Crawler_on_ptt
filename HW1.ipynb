```python
import datetime
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select

tens = [
        ('one',8),
        ('two',10),
        ('three',9),
        ('four',15),
        ('five',18),
        ('six',12),
        ('seven',13),
        ('eight',17),
        ('nine',20),
        ('ten',19)]
tenposts = []
class sort:
    
    def putin(self,h,n):
        p=(h,n)
        tens.append(p)
        tens.sort(reverse = True , key = lambda s:s[1])
        del tens[10]
    def out(self):
        print(tens)
    def post(self,t):
        for address in tens:
            url = address[0]
            driver.get(url)
            data = driver.find_elements_by_class_name('article-meta-value')
            poster = data[0].text
            title = data[2].text
            posttime = data[3].text
            maincontainer = driver.find_element_by_id('main-container')
            text = maincontainer.text
            ptext = text.split('--')[0]
            texts = ptext.split('\n')
            print(texts)
            fartical = '\n'.join(texts[4:])
            IPs = driver.find_elements_by_class_name('f2')
            for i in IPs:
                if '發信站' in i.text:
                    postIP = i.text
                    break
            final = [poster,title,posttime,postIP,fartical]
            tenposts.append(final)
        import csv
        with open('HW1.csv','w',newline='',encoding='utf-8-sig') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['作者','標題','時間','IP','內文'])
            for f in tenposts:
                writer.writerow(f)
            writer.writerow(["爬蟲日期",t])


options = Options()
options.binary_location = "C:\\Program Files (x86)\\Google\\Chrome\\Application\\chrome.exe"
webdriver_path = 'C:\\chrome.exe'
driver = webdriver.Chrome(executable_path = 'C:\\Users\\UX433\\Desktop\\Untitled Folder\\chromedriver.exe')
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
driver.get(url)
yes_btn = driver.find_element_by_name('yes')
yes_btn.click()
posts = driver.find_elements_by_class_name('r-ent')
today = datetime.date.today()
timetoday = today.strftime("%m/%d")
timerange = datetime.timedelta(days = 1)
yesterday = today - timerange
lastyesterday = yesterday - timerange
timeyesterday = yesterday.strftime("%m/%d")
timelastyesterday = lastyesterday.strftime("%m/%d")
datepost = [("",timetoday)]
for post in posts:
    d = post.find_element_by_class_name('date').text
    date = datetime.datetime.strptime(d,"%m/%d")
    if(date == timeyesterday ):
        try:
            s = post.find_element_by_tag_name('a')
        except :
            continue
        else:
            datepost.append((s.get_attribute('href'),date))


e = 0
lastpage_btn = driver.find_element_by_link_text('‹ 上頁')
lastpage_btn.click()
posts = driver.find_elements_by_class_name('r-ent')
date = posts[0].find_element_by_class_name('date').text
d = datetime.datetime.strptime(date,"%m/%d")
while(date >= timelastyesterday):
    for post in posts:
        di = post.find_element_by_class_name('date').text
        date = datetime.datetime.strptime(di,"%m/%d")
        if(di == timeyesterday):
            try:
                s = post.find_element_by_tag_name('a')
            except :
                continue
            else:
                datepost.append((s.get_attribute('href'),date))
    
    lastpage_btn = driver.find_element_by_link_text('‹ 上頁')
    lastpage_btn.click()
    posts = driver.find_elements_by_class_name('r-ent')
    date = posts[0].find_element_by_class_name('date').text
    d = datetime.datetime.strptime(date,"%m/%d")

s = sort()
for post in datepost:
    if (post == datepost[0]):
        continue
    u = post[0]
    url =u
    driver.get(url)
    pushs = driver.find_elements_by_class_name('push')
    n = len(pushs)
    s.putin(url,n)

s.post(timetoday)

driver.close()
```
